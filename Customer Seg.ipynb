```python
from sklearn.cluster import KMeans,AgglomerativeClustering
# from scipy.stats import zscore
```

```python
# df_CS_z = zscore(df_CS)
df_CS_N = df_CS.drop(columns=['CustomerID','Age','Gender_Female', 'Gender_Male'])
```

```python
df_CS_N
```

```python
df_CS_z = df_CS_N.apply(zscore)
```

```python
df_CS_N.head(2)
```

```python
cluster_range = range(1,10)
cluster_errors = []

for num_cluster in cluster_range:
    clusters = KMeans(num_cluster)
    clusters.fit(df_CS_z)
    labels = clusters.labels_
    centroids = clusters.cluster_centers_
    cluster_errors.append(clusters.inertia_)
clusters_df = pd.DataFrame({'num_of_cluster':cluster_range, 'cluster_errors': cluster_errors })
clusters_df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12,6))
plt.plot(clusters_df.num_of_cluster,clusters_df.cluster_errors, marker = 'o')
```

```python
# Agglomerative Clustering
```

```python
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean',  linkage='average')
```

```python
model.fit(df_CS_N)
```

```python
df_CS_N['labels'] = model.labels_
```

```python
df_CS_N.groupby('labels').count()
```

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
```

```python
from scipy.spatial.distance import pdist
```

```python
Z = linkage(df_CS_N,'ward')
c,coph_dist = cophenet(Z, pdist(df_CS_N))
```

```python
c
```

```python
plt.figure(figsize=(12,6))
plt.title('Agglomerative Hiercichal Distance')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold = 40, leaf_font_size=8.)
plt.tight_layout()
```

```python
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=5,  # show only the last p merged clusters
)
plt.show()
```
